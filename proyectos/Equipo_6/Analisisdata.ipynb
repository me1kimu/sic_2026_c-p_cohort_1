```python
%%writefile app.py
import streamlit as st
import pandas as pd
import datetime


df = pd.read_csv("clean_inventario.csv")


df['date'] = pd.to_datetime(df['date']).dt.date

st.sidebar.header("Filtros")

fecha_minima = df['date'].min()
fecha_maxima = df['date'].max()

rango_fechas = st.sidebar.date_input(
    "Selecciona el rango de fechas:",
    value=(fecha_minima, fecha_maxima),
    min_value=fecha_minima,
    max_value=fecha_maxima
)

if len(rango_fechas) == 2:
    inicio, fin = rango_fechas

    df_filtrado = df[(df['date'] >= inicio) & (df['date'] <= fin)]

    st.write(f"### Datos filtrados desde {inicio} hasta {fin}")
    st.dataframe(df_filtrado)
else:
    st.warning("Por favor, selecciona la fecha de fin en el calendario.")
```

```python
df = pd.read_csv('clean_inventario.csv')
df.to_sql('invt_sql', conn, index=False, if_exists='replace')
```

```python
df.columns = df.columns.str.strip()
```

```python
category_summary = df.groupby('category').agg({
    'units_sold': 'sum',
    'inventory_level': 'mean',
    'demand_forecast': 'mean'
}).reset_index()
print(category_summary)

print("\n--- MÉTRICAS POR REGIÓN ---")
region_summary = df.groupby('region').agg({
    'units_sold': 'sum'
}).reset_index()
print(region_summary)

pivot_sales = df.pivot_table(index='category', columns='region', values='units_sold', aggfunc='sum')

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_sales, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Ventas Totales por Categoría y Región")
plt.ylabel("Categoría de Producto")
plt.xlabel("Región de la Tienda")
plt.show()
```

```python
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(16, 5))


order_category = df.groupby('category')['inventory_level'].sum().sort_values(ascending=False).index

sns.barplot(
    data=df,
    x='category',
    y='inventory_level',
    order=order_category,
    estimator=sum,
    errorbar=None,
    ax=axes[0],
    palette='Blues_r'
)
axes[0].set_title('Total de Inventario por Categoría de Producto')
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45, ha='right')
axes[0].yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))


order_region = df.groupby('region')['inventory_level'].sum().sort_values(ascending=False).index

sns.barplot(
    data=df,
    x='region',
    y='inventory_level',
    order=order_region,
    estimator=sum,
    errorbar=None,
    ax=axes[1],
    palette='Oranges_r'
)
axes[1].set_title('Total de Inventario por Región Geográfica')
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45, ha='right')
axes[1].yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

plt.tight_layout()
plt.show()
```

Existe una diferencia que a escalas grandes parece mínima, donde se ve un comportamiento ligeramente decreciente en los datos alineados de esa forma específica.

```python


print("\n--- MÉTRICAS DE VENTAS E INVENTARIO POR CATEGORÍA ---")
category_analysis = df.groupby('category').agg(
    Ventas_Totales=('units_sold', 'sum'),
    Ventas_Promedio=('units_sold', 'mean'),
    Inventario_Promedio=('inventory_level', 'mean'),
    Pronostico_Demanda_Promedio=('demand_forecast', 'mean')
).reset_index()
print(category_analysis.to_string(index=False))

plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='category', y='units_sold', palette='Set2')
plt.title('Dispersión y Outliers de Unidades Vendidas por Categoría')
plt.xticks(rotation=30)
plt.show()
```

```python

pivot_sales = df.pivot_table(
    index='category',
    columns='region',
    values='units_sold',
    aggfunc='sum'
)

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_sales, annot=True, fmt=",d", cmap="YlGnBu", linewidths=.5)
plt.title('Unidades Vendidas por Categoría y Región')
plt.ylabel('Categoría de Producto')
plt.xlabel('Región de la Tienda')
plt.show()
```

```python

pivot_sales = df.pivot_table(
    index='category',
    columns='region',
    values='inventory_level',
    aggfunc='sum'
)

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_sales, annot=True, fmt=",d", cmap="YlGnBu", linewidths=.5)
plt.title('Nivel de Inventario por Categoría y Región')
plt.ylabel('Categoría de Producto')
plt.xlabel('Región de la Tienda')
plt.show()
```

Se observan diferentes objetivos de ventas segun las regiones de las tiendas, siguiendo con un ejemplo anterior existe una similitud de las unidades de venta al inventario

```python

pivot_weather = df.pivot_table(
    index='region',
    columns='weather_condition',
    values='units_sold',
    aggfunc='mean'
)

plt.figure(figsize=(10, 5))
sns.heatmap(pivot_weather, annot=True, fmt=".1f", cmap="magma", linewidths=.5)
plt.title('Ventas Promedio Diario por Región y Condición Climática')
plt.ylabel('Región')
plt.xlabel('Condición Climática')
plt.show()
```

En general no hay mucha diferencia, pero se puede apreciar que los dias soleados, y en especial en el este, se venden más articulos por dia
