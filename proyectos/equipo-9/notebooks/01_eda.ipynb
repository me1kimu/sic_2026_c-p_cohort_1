Vamos a verificar la relación entre `is_returned` y `return_reason`:

```python
# Verificar si hay valores NaN en 'return_reason' cuando 'is_returned' es True
null_reasons_for_returned = df[df['is_returned'] == True]['return_reason'].isnull().sum()
print(f"Número de 'return_reason' nulos cuando 'is_returned' es True: {null_reasons_for_returned}")
```

```python
# Verificar si hay valores no NaN en 'return_reason' cuando 'is_returned' es False
not_null_reasons_for_not_returned = df[df['is_returned'] == False]['return_reason'].notnull().sum()
print(f"Número de 'return_reason' no nulos cuando 'is_returned' es False: {not_null_reasons_for_not_returned}")

if null_reasons_for_returned == 0 and not_null_reasons_for_not_returned == 0:
    print("\nLa hipótesis es correcta: 'return_reason' tiene un valor solo si 'is_returned' es True, y es NaN si 'is_returned' es False.")
else:
    print("\nLa hipótesis no es completamente correcta. Revisa los conteos anteriores.")
```

```python
df['return_reason'] = df['return_reason'].fillna('no_returned')

# Opcional: Verificar los cambios mostrando el conteo de valores de nuevo
print("Conteo de valores de 'return_reason' después de reemplazar NaN:")
display(df['return_reason'].value_counts())
```

```python
# Opcional: Verificar que no queden NaNs en la columna
print("Número de NaNs en 'return_reason' después de la operación:")
display(df['return_reason'].isnull().sum())
```
